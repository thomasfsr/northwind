```python
import pandas as pd
import psycopg2 as pg
pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python
host = "localhost"
database = "northwind"
user = "postgres"
password = "postgres"
port = 5432
```

```python
conn = pg.connect(host=host, port=5432, dbname=database, user=user, password=password)
cur = conn.cursor()
```

```python
tables = [
    'categories',
    'customers',
    'employee_territories',
    'employees',
    'order_details',
    'orders',
    'products',
    'region',
    'shippers',
    'suppliers',
    'territories',
    'us_states'
]

dfs = {}
for table in tables:
    with conn as con:
        cur = con.cursor()
        cur.execute(f'SELECT * FROM {table}')
        rows = cur.fetchall()
        columns = [desc[0] for desc in cur.description]
        dfs[table] = pd.DataFrame(rows, columns=columns)
```

```python
cat = dfs['categories']
cus= dfs['customers']
et= dfs['employee_territories']
emp = dfs['employees']
od= dfs['order_details']
ord = dfs['orders']
pro=dfs['products']
reg=dfs['region']
shi=dfs['shippers']
sup=dfs['suppliers']
ter=dfs['territories']
us=dfs['us_states']
```

```python
od_by_order_id= od[['order_id',
                    'quantity',
                    'raw_price',
                    'price',
                    'discount',
                    'quantity']].groupby('order_id', as_index=False).sum()
```

```python
int_ord = pd.merge(ord, od_by_order_id, how='left' ,on = 'order_id')
```

```python
cat_cols = ['category_id','category_name','category_description']
pro_join_cat = pd.merge(pro, cat[cat_cols], on='category_id', how='left')
```
